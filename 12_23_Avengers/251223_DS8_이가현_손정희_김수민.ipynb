## [QUEST 02: Avengers vocab 만들기]

### 1. Vocabulary란

- 단어가 특정 정수와 매칭되어 있는 딕셔너리 형태의 단어장

### 2. 퀘스트 목표

- 문장을 입력받아서 정수로 리턴해주는 함수를 만들자!

### 3. 수행 조건

1. **텍스트 전처리를 수행한다.**
    - `Avengers.txt` 파일을 불러온다.
    - 모든 문자는 **소문자**로 변환한다.
    - 모든 **기호**를 제거한다.
    - **단어 단위**로 분리한다.
    - **결과 예시:** `"I am hungry..very much..."` -> `['i', 'am', 'hungry', 'very', 'much']`
2. 각 단어별 **빈도수**를 딕셔너리 형태로 저장한다.
3. 딕셔너리를 **빈도수 순으로 내림차순 정렬**하고, 정렬 순서대로 **정수 인덱스**를 부여한다.
4. 텍스트를 `input()`으로 입력받아서 정수를 `return`하는 함수(`encoder`)를 만든다.

```python
#===== 0. 필요모듈 임포트=====

# re 모듈을 불러옴, re는 "regular expression"(정규표현식)의 약자
# 문자열에서 패턴을 찾거나, 매칭하거나, 치환하는 작업에 사용
import re
# collections 모듈에서 Counter 클래스만 불러옴
# Counter는 해시 가능한(hashable) 객체의 개수를 세는 딕셔너리 서브클래스
from collections import Counter
```

```python
#===== 1. 텍스트 천처리=====

# 1-1. 파일 불러오기
# open() 함수로 파일을 읽기 모드('r')로 열기, f는 파일 객체를 참조하는 변수
with open('/content/Avengers.txt', 'r') as f:       # 파일 경로 통일 필요
  # read() 메서드로 파일의 전체 내용을 하나의 문자열로 읽어 data 변수에 저장
  data =  f.read()

# 1-2. 소문자로 통일
# lower() 메서드로 모든 대문자를 소문자로 변환
# 이유: 'Apple'과 'apple'을 같은 단어로 취급하기 위함
data = data.lower()

# 1-3. 모든 기호 제거(알파벳, 공백만 남김)
# isalnum(): 알파벳 또는 숫자인지 확인 (True/False 반환)
# x == ' ': 공백 문자인지 확인
# 둘 중 하나라도 True면 해당 문자를 리스트에 포함
data = [x for x in data if x.isalnum() or x == ' ']
data = ''.join(data)

# 1-4. 단어단위 분리
# 공백() 을 기준으로 문자열을 분리하여 단어 리스트 생성
words = data.split()

# 결과 확인
print(words)
```

```python
# ===== 2. 각 단어별 빈도수를 딕셔너리 형태로 저장 =====
# Counter 객체 생성: 리스트의 각 요소(단어) 출현 횟수를 자동 집계
words_count = Counter(words)

# 결과 확인
print(words_count)
```

```python
# ===== 3. 딕셔너리를 빈도수 순으로 내림차순 정렬하고, 정렬 순서대로 정수 인덱스를 부여 =====
# 3-1. 딕셔너리를 빈도수 순으로 내림차순 정렬
sorted_words_count = sorted(words_count.items(), key=lambda x: x[1], reverse=True)

# 3-2. 정렬 순서대로 정수 인덱스를 부여
word_to_index = {word: index for index, (word, count) in enumerate(sorted_words_count)}

# 결과 확인
print(word_to_index)
```

```python
# ===== 4. 텍스트를 input()으로 입력받아서 정수를 return하는 함수 =====
# 4-1. 단어들 인코딩하기
def encoder(text):
  text = text.lower()   # 소문자 변환  (상단 코드 블럭 참조)

  text = [x for x in text if x.isalnum() or x == ' ']
  text = ''.join(text)  # 알파벳과 공백 외에 제거 (상단 코드 블럭 참조)

  words = text.split()  # 공백 기준으로 단어분리 (상단 코드 블럭 참조)

  result = []           # 숫자로 바뀐 리스트로 결과 도출

  for w in words:                         # 입력한 단어를 하나씩 꺼낸다
    if w in word_to_index:            # 단어가 딕셔너리에 존재하는지 확인
      result.append(word_to_index[w]) # 단어가 딕셔너리에 존재하면 해당 단어 인덱스를 리스트에 추가

  return result  # 인덱스 리스트 반환


#결과 확인
text = input()
encoder(text)
```

```python
# 전처리 함수
def preprocess(text):
  text = text.lower()
  # text = [x for x in text if x.isalnum() or x == ' ']

  text = re.sub(r'[^a-z0-9 ]', '', text) # 정규표현식

  text = ''.join(text)
  words = text.split()
  return words


with open('/content/Avengers.txt', 'r') as f:
  data = f.read()

words = preprocess(data)

words_count = Counter(words)
sorted_words_count = sorted(words_count.items(), key=lambda x: x[1], reverse=True)
word_to_index = {word: index for index, (word, count) in enumerate(sorted_words_count)}


def encoder(text):
  words = preprocess(text)
  # 리스트 컴프리헨션 -> 한줄로 간단하게 표현.
  return [word_to_index[w] for w in words if w in word_to_index]


# 입력
text = input()
print(encoder(text))
```
